````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]


# Step 1: Define the function for initializing the inventory with error handling:


def initialize_inventory(products):
    print("\nInventory")
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter the available quantity for {product}: "))
                if quantity < 0:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
                    continue
                inventory[product] = quantity
                break
            except ValueError:
                print("Invalid input. Please enter an integer quantity.")
    return inventory

    



def get_customer_orders():
    print("\nCustomer order:")

    num_orders = int(input("How many products does the customer want to order? "))

    # list:
    choice_orders = [
        input(f"Enter the name of product #{i + 1}: ").strip().lower()
        for i in range(num_orders)
    ]
 
    customer_orders = {product for product in choice_orders if product in products}

    return customer_orders


# Step 2. Modify the `calculate_total_price` function to include error handling:

def get_customer_orders(inventory):
    print("\nCustomer order:")

    while True:
        try:
            num_orders = int(input("How many products does the customer want to order? "))
            if num_orders < 0:
                print("Number of products cannot be negative. Please enter a valid number.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    choice_orders = []

    for i in range(num_orders):
        while True:
            product = input(f"Enter the name of product #{i + 1}: ").strip().lower()

            if product not in products:
                print("Unknown product. Please enter a valid product name.")
                print(f"Available products: {', '.join(products)}")
                continue

            if inventory.get(product, 0) <= 0:
                print(f"No stock available for '{product}'. Please choose another product.")
                continue

            choice_orders.append(product)
            break

    customer_orders = {product for product in choice_orders if product in products}
    return customer_orders





def update_inventory(customer_orders, inventory):
    updated_inventory = {
        product: (quantity - 1 if product in customer_orders and quantity > 0 else quantity)
        for product, quantity in inventory.items()
    }

    updated_inventory = {
        product: quantity
        for product, quantity in updated_inventory.items()
        if quantity > 0
    }

    return updated_inventory



def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered


def print_order_statistics(order_statistics):
    total_products_order, percentage_order = order_statistics

    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_order}")
    print(f"Percentage of Products Ordered: {percentage_order:}")



def print_updated_inventory(inventory):
    print("\nUpdated Inventory")

    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]




# Stp 3. Modify the `get_customer_orders` function to include error handling:

def calculate_total_price(customer_orders):
    print("\nPrice input for each ordered product:")

    total_price = 0.0

    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price for '{product}': "))
                if price < 0:
                    print("Price cannot be negative. Please enter a valid price.")
                    continue
                total_price += price
                break
            except ValueError:
                print("Invalid input. Please enter a numeric price.")

    return total_price



def total_price_order():
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders(inventory)

    order_stats = calculate_order_statistics(customer_orders, products)

    total_price = calculate_total_price(customer_orders)

    inventory = update_inventory(customer_orders, inventory)

    print_order_statistics(order_stats)
    print_updated_inventory(inventory)

    print(f"\nTotal price of the customer order: {total_price:.1f}")


if __name__ == "__main__":
    total_price_order()
````
